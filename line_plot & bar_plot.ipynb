```python
import pandas as pd

# فرض میکنیم قیمت ها در دیکشنری بصورت زیر قرار دارد
dct_price = {
    "Samsung A34"                 : [14_000_000, 13_000_000, 
                                     12_000_000, 10_800_000],
    
    "Xiaomi Redmi Note 12"        : [10_500_000, 9_500_000 , 
                                     8_700_000 , 7_790_000],
    
    "Apple iPhone 13"             : [50_000_000, 40_000_000, 
                                     37_000_000, 34_500_000],
    
    "Samsung Galaxy S23 Ultra 5G" : [55_000_000, 52_500_000, 
                                     50_000_000, 48_000_000]
}

df = pd.DataFrame.from_dict(
    dct_price, orient="index", 
    columns=['Spring', 'Summer', 'Fall', 'Winter']
).reset_index(names = 'Phone model')

df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df = df.melt('Phone model', var_name='Season', 
             value_name='Price')

sns.catplot(x="Season", y="Price", hue='Phone model', 
            data=df, kind='point')

plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.barplot(x="Season", y="Price", 
            hue='Phone model', data=df)

plt.ticklabel_format(style='plain', axis='y')
plt.show()
```
